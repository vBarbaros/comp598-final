```python
import json
import numpy as np
from twitter_stats_utils import read_from_json, plot_stats_graph, plot_stats_bar

dict_stats = read_from_json('data/results/stats_per_covid_label.json')
print(json.dumps(dict_stats, indent=4))
```

```python
data = [[dict_stats['covid-controversy']['sentiment']['pos'][0],
         dict_stats['covid-new-variant']['sentiment']['pos'][0],
         dict_stats['pandemic-legal']['sentiment']['pos'][0],
         dict_stats['pandemic-logistics']['sentiment']['pos'][0],
         dict_stats['vax-efficiency']['sentiment']['pos'][0]],
        
        
        [dict_stats['covid-controversy']['sentiment']['neg'][0],
         dict_stats['covid-new-variant']['sentiment']['neg'][0],
         dict_stats['pandemic-legal']['sentiment']['neg'][0],
         dict_stats['pandemic-logistics']['sentiment']['neg'][0],
         dict_stats['vax-efficiency']['sentiment']['neg'][0]],
        
        [dict_stats['covid-controversy']['sentiment']['neut'][0],
         dict_stats['covid-new-variant']['sentiment']['neut'][0],
         dict_stats['pandemic-legal']['sentiment']['neut'][0],
         dict_stats['pandemic-logistics']['sentiment']['neut'][0],
         dict_stats['vax-efficiency']['sentiment']['neut'][0]]

       ]

print(data)

X = np.arange(5)
title = 'Sentiment values per COVID-specific Labels'
labels = ['pos', 'neg', 'neut']
xlabel = 'COVID label'
ylabel = 'Nr. of tweets'
xticks = ['controversy', 'new-variant', 'legal', 'logistics', 'efficiency']

plot_stats_bar(data, labels, xlabel, ylabel, title, xticks)
```
